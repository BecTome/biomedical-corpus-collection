```python
from datasets import load_dataset
import pandas as pd

# Login using e.g. `huggingface-cli login` to access this dataset
ds = load_dataset("chizhikchi/CARES")
```

```python
df_train = ds["train"].to_pandas()
df_test = ds["test"].to_pandas()

print("TRAIN")
print(df_train.head().to_markdown())
print()
print("TEST")
print(df_test.head().to_markdown())
print()
print("Training size:", len(df_train))
print("Test size:", len(df_test))

df = pd.concat([df_train, df_test], ignore_index=True)

print("Total size:", len(df))
```

```python
import os
if not os.path.exists("data/CARES"):
    os.mkdir("data/CARES")
df.to_csv("data/CARES/data.csv", index=False)
```

```python
from src.utils import count_tokens, count_sentences
import swifter

df["tokens"] = df["full_text"].swifter.apply(count_tokens)
df["sentences"] = df["full_text"].swifter.apply(count_sentences)
```
